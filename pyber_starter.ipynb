```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib import colors as mcolors
import cufflinks as cf
import plotly.graph_objs as go


# File to Load (Remember to change these)
city_data = pd.read_csv("city_data.csv")
ride_data = pd.read_csv("ride_data.csv")


# Read the City and Ride Data
#print(city_data.head())
#print(ride_data.head())

# Combine the data into a single dataset
df = pd.merge(ride_data, city_data, how="left", on=("city", "city"))
# Extract the year
df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].dt.year

# Display the data table for preview
df.head()
```

```python
# Create a datframe for metrics: Average Fare ($) Per City, # of Rides Per City, # of Drivers Per City, City Type
#df["type"].unique()
types_df = df.groupby(['city', 'type', 'driver_count']).agg({'fare':'mean','ride_id': 'count', 'driver_count': 'sum'}).rename(columns = {'fare': 'Avrg Fare', 'ride_id': 'Total Rides', 'driver_count': 'Total Drivers'}).reset_index()

types_df.head()
```

```python
# Create different dataframes for each type

urban = types_df.loc[types_df["type"] == "Urban"]
rural = types_df.loc[types_df["type"] == "Rural"]
suburban = types_df.loc[types_df["type"] == "Suburban"]
```

## Bubble Plot of Ride Sharing Data

```python
from matplotlib.lines import Line2D

# Build the scatter plots for each city types
plt.scatter(urban["Total Rides"], urban["Avrg Fare"],
            s=urban["driver_count"] *10.5, alpha=.7,color = 'lightcoral', edgecolors='black');

plt.scatter(rural["Total Rides"], rural["Avrg Fare"],
            s=rural["driver_count"] *10.5, alpha=.7,color = 'gold', edgecolors='black');

plt.scatter(suburban["Total Rides"], suburban["Avrg Fare"],
            s=suburban["driver_count"] *10.5, alpha=.7,color = 'lightblue', edgecolors='black');

# Incorporate the other graph properties
title_font = {'fontname':'Arial', 'size':'16', 'color':'black', 'weight':'normal',
              'verticalalignment':'bottom'} 
axis_font = {'fontname':'Arial', 'size':'14'}

plt.xlabel('Total Number of Rides (per city)', **axis_font);
plt.ylabel('Average Fare ($)', **axis_font);
plt.title('Pyber Ride Sharing Data (2018)', **title_font);
plt.style.use('seaborn-whitegrid')

# Create a legend
line1 = Line2D(range(1), range(1), color="white", marker='o', markersize=10, markerfacecolor="lightcoral")
line2 = Line2D(range(1), range(1), color="white", marker='o',markersize=10, markerfacecolor="gold")
line3 = Line2D(range(1), range(1), color="white", marker='o',markersize=10, markerfacecolor="lightblue")
plt.legend((line1,line2,line3),('Urban','Rural', 'Suburban'),numpoints=1, loc=1, title='City Type', frameon=True);

# Incorporate a text label regarding circle size
plt.text(42, 40, 'Note:\nCircle size correlates with driver sizer per city', fontsize=14);


# Save Figure
plt.savefig("Pyber_BubblePlot.png")
```

## Total Fares by City Type

```python
# Calculate Type Percents
CityTypesFares = df.groupby(['type']).agg({'fare': 'sum'}).reset_index()

# Build Pie Chart
labels = types_df['type']
colors = ["gold", "lightblue", "lightcoral"]
explode = (0, 0, 0.1)

CityTypesFares.fare.plot(kind='pie', colors=colors, labels=labels, autopct="%1.1f%%", explode=explode, startangle=150, 
                   title="% of Total Fares by City Type", shadow=True);
plt.ylabel('')

# Save Figure
plt.savefig("City_Type_Fares.png")
```

## Total Rides by City Type

```python
# Calculate Ride Percents
CityTypeRides = df.groupby(['type']).agg({'ride_id': 'count'}).reset_index()

# Build Pie Chart
labels = types_df['type']
colors = ["gold", "lightblue", "lightcoral"]
explode = (0, 0, 0.1)

CityTypeRides.ride_id.plot(kind='pie', colors=colors, labels=labels, autopct="%1.1f%%", explode=explode, startangle=150, 
                   title="% of Total Rides by City Type", shadow=True);
plt.ylabel('')

# Save Figure
plt.savefig("City_Type_Rides.png")
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
driver_percent = city_data.groupby(['type']).agg({'driver_count': 'sum'}).reset_index()

# Build Pie Charts
labels = driver_percent['type']
colors = ["gold", "lightblue", "lightcoral"]
explode = (0, 0, 0.1)

driver_percent.driver_count.plot(kind='pie', colors=colors, labels=labels, autopct="%1.1f%%", explode=explode, startangle=180, 
                   title="% of Total Drivers by City Type", shadow=True);
plt.ylabel('')

# Save Figure
plt.savefig("City_Type_Drivers.png")
```
